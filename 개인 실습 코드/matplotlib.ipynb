```python
# import numpy
import numpy as np
# linspace 함수를 이용해서 선 그리기( -2, 2 범위에 속하는 부동소수로 구성된 x의 배열을 생성)
x = np.linspace(-2, 2, 500)

print(x)

y = x**2
plt.plot(x, y)
plt.show()
```

```python
# 위 그래프에 속성 추가
plt.plot(x, y)
plt.title("Square function == 이차함수") # 타이틀
plt.xlabel("x") # x축 레이블
plt.ylabel("y == x**2") # y축 레이블
plt.grid(True) # 격자점 추가
plt.show()# 출력
```

## matplotlib in python LINE DRAWING ##

```python
# 바로 다음에 위치한 데이터 사이에 선을 그려준다.
plt.plot(
    [0, 100, 100, 0 , 0, 100, 50, 0, 100], # x좌표
    [0, 0, 100, 100, 0 ,100, 130, 100, 0]) # y좌표
plt.axis([-10, 110, -10, 140])
plt.show()
```

```python
# line custom 1
plt.plot(
    [0, 100, 100, 0 , 0, 100, 50, 0, 100],
    [0, 0, 100, 100, 0 ,100, 130, 100, 0],
    "g--") # green clolor && -- line type
plt.axis([-10, 110, -10, 140])
plt.show()
```

```python
# line custom 2
plt.plot(
    [0, 100, 100, 0, 0], [0, 0, 100, 100, 0],"r-",[0, 100, 50, 0, 100], [0, 100, 130, 100, 0],
    "g--")
plt.axis([-10, 110, -10, 140])
plt.show()
```

```python
# plt에서는 show 이전까지 속성 여러번 부여 가능
# line custom 2 분할 코딩
plt.plot([0, 100, 100, 0, 0],
        [0, 0, 100, 100, 0], "r-")      
plt.plot([0, 100, 50, 0, 100],
        [0, 100, 130, 100, 0], "g--")
plt.axis([-10, 110, -10, 140])

# 원하는 속성을 3번 나눠서 부여한 이후 plt.show()
plt.show()
```

```python
# 여러 그레프 표현
x = np.linspace(-14, 14, 100)
plt.plot(x, x, 'g--')
plt.plot(x, x**2, 'r')
plt.plot(x, x**3, 'b^--')
plt.axis([-10, 10, -10, 10])

plt.show()
```

```python
# 한번에 속성값 입력 후 분할
# plot는 자동으로 LINE2D객체의 리스트를 반환함(각 객체는 선에 대응)
x = np.linspace(-1.4, 1.4, 30)
line1, line2, line3 = plt.plot(x, x, 'g--', x, x**2, 'r', x, x**3, 'b^--')

print(type(line1)) # 각 선의 type은 Line2D

# 각각의 Line Custom 가능
line1.set_linewidth(3.0) #line1의 선 넓이(굵기)
line1.set_dash_capstyle("round") # 선 cap style(선 끝 모양)
line3.set_alpha(0.2) # 선 투명도
plt.show()
```

## Saving Plot ##

```python
# Savefig 함수를 이용해서 저장한다.
x = np.linspace(-1.4, 1.4, 30)
plt.plot(x,x**2)

plt.savefig("plotname.png", transparent=True) #transparent = True (png 파일에서 배경을 투명하게 사용하겠다는 옵션)

# 앞에서와 마찬가지로 저장 함수도 plt.show() 이전에 실행해야함(plt.show() -> 모든 함수(or 옵션) 실행 후 Clear)
```
