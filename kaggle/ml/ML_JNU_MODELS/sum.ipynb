# Merge to One CSV File Code #

## 이 코드는 전처리 과정이 다른 data를 사용한 경우, 기존 Sklearn에서 제공하는 Voting을 이용할 수 없기 때문에, 수작업으로 하드코딩한 작업물 입니다. ##

```python
import pandas as pd
import numpy as np
import sklearn
import tensorflow as tf
```

```python
r1 = pd.read_csv("r1.csv")
r2 = pd.read_csv("r2.csv")
r3 = pd.read_csv("r3.csv")
r4 = pd.read_csv("r4.csv")


from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse=False)
r1_arr = ohe.fit_transform(r1[['method']])
r2_arr = ohe.fit_transform(r2[['method']])
r3_arr = ohe.fit_transform(r3[['method']])
r4_arr = ohe.fit_transform(r4[['method']])
```

```python
arr = r1_arr + r2_arr +r3_arr + r4_arr
```

```python
print(arr)
```

```python
a = []
```

```python
for i in range(285):
    a.append(arr[i].argmax())
```

```python
arra = np.array(a)
```

```python
pds = pd.Series(arra).copy()
```

```python
pds.to_csv('predict.csv',index=True)
```
